```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# %matplotlib inline
import seaborn as sns

df = pd.read_csv('/Users/leejieun/kaggle-notebooks/datasets/lsind18/euro-exchange-daily-rates-19992020/euro-daily-hist_1999_2022.csv')
```

```python
df
```

```python
df.rename(columns= {'[Swiss franc ]': 'CHF', 'Period\\Unit:': 'Time'}, inplace=True)
df['Time'] = pd.to_datetime(df['Time'])
df.sort_values('Time', inplace=True)
df.reset_index(drop=True, inplace=True)

eur_chf = df[['Time', 'CHF']]
eur_chf = eur_chf[eur_chf['CHF'] != '-']
eur_chf['CHF'] = eur_chf['CHF'].astype(float)
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (17, 5))

plt.plot(eur_chf['Time'], eur_chf['CHF'], color = 'grey', alpha = 0.30)
plt.plot(eur_chf['Time'][3247:4107], eur_chf['CHF'][3247:4107], color = 'firebrick', linewidth = 2.5)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.tick_params(bottom = False, left = False)
for location in ['left', 'right', 'top', 'bottom']:
    ax.spines[location].set_visible(False)

ax.text(10200, 1.85, 'EURCHF DAILY CHART - SOMETHING LOOKS OUT OF PLACE', fontsize = 15, weight = 'bold')
ax.text(10400, 1.70, 'EUR-RELEASE', fontsize = 11, color = 'grey', alpha = 0.35)
ax.text(10400, 1.65, 'JAN-99', fontsize = 11, color = 'grey', alpha = 0.35)
ax.text(13600, 1.71, 'OCT-07', fontsize = 11, color = 'grey', alpha = 0.35)
ax.text(15040, 0.98, 'SEP-11', fontsize = 11, color = 'firebrick', weight = 'bold')
ax.text(16250, 0.91, 'JAN-15', fontsize = 11, color = 'firebrick', weight = 'bold')
ax.text(19150, 0.88,'NOV-22', fontsize = 11, color = 'grey', alpha = 0.35)

ax.axhline(0.80, linewidth = 17, color = 'grey')
ax.axhline(1.96, linewidth = 6, color = 'grey')

ax.text(10200, 0.778, '©Shaun Oilund' + ' '*154 + 'Source: European Central Bank',  color = 'white', size = 13)

plt.show()
```
